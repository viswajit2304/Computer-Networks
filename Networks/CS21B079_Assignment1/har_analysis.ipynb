```python
import nest_asyncio
nest_asyncio.apply()
import json
from haralyzer import HarParser

def q2(file_name):
    har_file_path = file_name

    with open(har_file_path, 'r', encoding='utf-8') as har_file:
        har_data = json.load(har_file)

    har_parser = HarParser(har_data)

    ii = 0
    for page in har_parser.pages:
        if page.page_id == 'page_1':
            page_load_time = page.page_load_time
            ttfb = page.time_to_first_byte
    count_get=0
    for page in har_parser.pages:
        for entry in page.entries:
            if entry['request']['method'] == 'GET':
                count_get=count_get+1

    mime_types_set = set()

    # Iterate through entries in the HAR file
    for page in har_parser.pages:
        for entry in page.entries:
            response = entry['response']
            content_type = response['content']['mimeType']
            mime_types_set.add(content_type)

    image_requests = 0
    javascript_requests = 0
    html_css_requests = 0
    for page in har_parser.pages:
        # Access entries for each page
        for entry in page.entries:
            response = entry['response']
            content_type = response['content']['mimeType']
            if content_type.startswith('image/'):
                image_requests += 1
            elif content_type == 'application/javascript':
                javascript_requests += 1
            elif content_type.startswith('text/html') or content_type.startswith('text/css'):
                html_css_requests += 1
    print(f"Fraction of GET requests for Images: {image_requests/count_get}")
    print(f"Fraction of GET requests for JavaScript: {javascript_requests/count_get}")
    print(f"Fraction of GET requests for HTML and CSS: {html_css_requests/count_get}")

    total_assets_size = 0
    total_images_size = 0
    image_sizes = []

    for page in har_parser.pages:
        # Access entries for each page
        for entry in page.entries:
            response = entry['response']
            content_type = response['content']['mimeType']
            response_size = int(response['content']['size'])
            total_assets_size += response_size
            if content_type.startswith('image/'):
                total_images_size += response_size
                image_sizes.append(response_size)

    if total_assets_size > 0:
        fraction_images_size = total_images_size / total_assets_size
    else:
        fraction_images_size = 0.0
    mean_image_size = 0
    median_image_size = 0
    if image_sizes:
        mean_image_size = sum(image_sizes) / len(image_sizes)
        sorted_image_sizes = sorted(image_sizes)
        median_image_size = sorted_image_sizes[len(sorted_image_sizes) // 2]

    for mime_type in mime_types_set:
        print(mime_type)
    print(" ")
    print(f"Total size of assets downloaded: {total_assets_size} bytes")
    print(f"Fraction of size constituted by images: {fraction_images_size}")
    print(f"Mean size of images: {mean_image_size} bytes")
    print(f"Median size of images: {median_image_size} bytes")
    print(" ")
    print(f"total_get = {count_get}")
    print(" ")
    print(f"TTFB: {ttfb} ms")
    print(f"Page Load Time: {page_load_time} ms")
    return
```

```python
print("Form deccan har file analysis :: ")
q2('deccan.har')
print("\n")
print("Form jagron har file analysis :: ")
q2('jagran.har')
print("\n")
print("Form mit har file analysis :: ")
q2('mit.har')
print("\n")
print("Form sinu har file analysis :: ")
q2('sinu.har')
print("\n")
print("Form usach har file analysis :: ")
q2('usach.har')
```
